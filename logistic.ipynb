```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button
```

```python
init_r = 1 + 2 * np.sqrt(2)

N = 500
x = 0.5 + np.zeros(N)

for n in range(N-1):
    x[n + 1] = init_r * x[n] * (1 - x[n])
```

```python
%matplotlib notebook

fig, ax = plt.subplots()
plt.ylim(0,1)
line, = plt.plot(x)

plt.subplots_adjust(bottom = 0.25)

ax_r = plt.axes([0.12, 0.1, 0.65, 0.03])
r_slider = Slider(
    ax = ax_r,
    label = 'r',
    valmin = 0,
    valmax = 4,
    valinit = init_r,
)

def update(val):
    for n in range(N-1):
        x[n + 1] = r_slider.val * x[n] * (1 - x[n])
    line.set_ydata(x)
    fig.canvas.draw_idle()
    
r_slider.on_changed(update)

plt.show()
```
